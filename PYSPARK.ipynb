```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# 1️⃣ Create Spark session
spark = SparkSession.builder \
    .appName("SmallPySparkExample") \
    .getOrCreate()

# 2️⃣ Create sample data
data = [
    (1, "Alice", 29),
    (2, "Bob", 31),
    (3, "Charlie", 25),
    (4, "David", 35)
]

columns = ["id", "name", "age"]

df = spark.createDataFrame(data, columns)

# 3️⃣ Simple transformation: filter age > 30
filtered_df = df.filter(col("age") > 30)

# 4️⃣ Show results
print("=== Original Data ===")
df.show()

print("=== Filtered Data (age > 30) ===")
filtered_df.show()

# Stop Spark session
spark.stop()
```
