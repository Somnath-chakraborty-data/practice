```python
import logging

# Configure logging
logging.basicConfig(
    level=logging.INFO,  # Set log level
    format="%(asctime)s - %(name)s - %(levelname)s - %(message)s"
)

logger = logging.getLogger(__name__)

# Example usage
logger.debug("This is a debug message (for developers).")
logger.info("This is an info message (general updates).")
logger.warning("This is a warning message (something might be wrong).")
logger.error("This is an error message (something failed).")
logger.critical("This is a critical message (system crash imminent).")
```
