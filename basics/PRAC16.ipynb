```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
import os
```

```python
spark = SparkSession.builder.appName("Basics").master("local[*]").getOrCreate()
```

```python
data=[(10 ,'Anil',50000, 18),
(11 ,'Vikas',75000,  16),
(12 ,'Nisha',40000,  18),
(13 ,'Nidhi',60000,  17),
(14 ,'Priya',80000,  18),
(15 ,'Mohit',45000,  18),
(16 ,'Rajesh',90000, 10),
(17 ,'Raman',55000, 16),
(18 ,'Sam',65000,   17),
(15 ,'Mohit',45000,  18),
(13 ,'Nidhi',60000,  17),      
(14 ,'Priya',90000,  18),  
(18 ,'Sam',65000,   17)
     ]
```

```python
schema =['id','name','salary','manager_id']
```

```python
df = spark.createDataFrame(data,schema)
```

```python
df.show()
```

```python
df1 = df.distinct()
df1.show()
```

```python
df.count()
```

```python
df1.count()
```

```python
df.select("id","name","salary","manager_id").distinct().show()
```

```python
df2 = df.drop_duplicates(["id","name","salary","manager_id"])
df2.show()
```

```python
df2.count()
```

```python
df2.sort("salary").show()
```

```python
df2.sort(col("salary").desc()).show()
```

```python
df2.sort(col("salary").desc(),col("name").desc()).show()
```

```python
df2.select(sum("salary"),max("salary"),min("salary")).show()
```

```python
df.groupBy("manager_id")\
    .agg(sum("salary").alias("total_salary")).show()
```
