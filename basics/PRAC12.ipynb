```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
import os
```

```python
spark = SparkSession.builder\
    .appName("Practice")\
    .master("local[*]")\
    .getOrCreate()
```

DATAFRAME MODE

```python
data = [(1, "Luca"), (2, "Aiko"), (3, "Carlos"), (4, "Sofia"), (5, "Ethan")]
```

```python
schema =["id", "name"]
```

```python
df = spark.createDataFrame(data, schema)
```

```python
df.select("*").show()
```

```python
df.select("name").show()
```

```python
df.select(df.name).show()
```

```python
from pyspark.sql import Row 
new_row = Row(id=6, name="James")
```

```python
new_df = spark.createDataFrame([new_row])

# 3. Combine the original 'df' with the 'new_df'
df_combined = df.union(new_df)

# 4. Show the result
df_combined.show()
```

```python
df_combined.select(col("name")).show()
```
