```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
import os
```

```python
spark = SparkSession.builder\
    .appName("Practice")\
    .master("local[*]")\
    .getOrCreate()
```

```python
my_schema = ("match_id INT, season INT, city STRING, date STRING, team1 STRING, team2 STRING, toss_winner STRING, toss_decision STRING, result STRING, dl_applied INT, winner STRING, win_by_runs INT, win_by_wickets INT, player_of_match STRING, venue STRING, umpire1 STRING, umpire2 STRING, umpire3 STRING")
```

```python
file_path = "/home/somnath/my_vscode_project/input_data/ipl.csv"
```

READ MODE

```python
df = spark.read \
    .format("csv") \
    .option("header", "false") \
    .option("inferSchema", "false") \
    .schema(my_schema) \
    .option("mode", "PERMISSIVE") \
    .load(file_path)
```
